```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df.info()
df['job_skills']=df['job_skills'].apply(lambda salary: ast.literal_eval(salary) if pd.notna(salary) else salary)
```

```python
df_dataAnalyst=df[df['job_title_short']=='Data Analyst'].copy()
df_dataAnalyst
```

```python
df_exploded=df_dataAnalyst.explode('job_skills')
df_exploded
```

```python
skill_explode=df_exploded.groupby('job_skills').agg(
    salary_median=('salary_year_avg','median'),
    skill_count=('job_skills','count')
)
skill_explode=skill_explode.sort_values(by='skill_count',ascending=False).head(20)
skill_explode.reset_index(inplace=True)
```

```python
#Works Only If Index Is reset
skill_explode.plot(kind='scatter',x='skill_count',y='salary_median')
for salary,count,title in zip(skill_explode['salary_median'],skill_explode['skill_count'],skill_explode['job_skills']):
    plt.text(count,salary,title)
```

```python
#solution given on video 
skill_explode.plot(kind='scatter',x='skill_count',y='salary_median')
for i,txt in enumerate(skill_explode['job_skills']):
    plt.text(skill_explode['skill_count'].iloc[i],skill_explode['salary_median'].iloc[i],txt)
```
