```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df.info()
df['job_skills']=df['job_skills'].apply(lambda salary: ast.literal_eval(salary) if pd.notna(salary) else salary)
```

```python
df_DA_ind=df[df['job_country']=='India'].copy()
df_DA_ind['salary_year_avg'].plot(kind='hist',bins=20,edgecolor='black')
plt.xlim(0,300000)
ax=plt.gca()
ax.xaxis.set_major_formatter(lambda x,pos : f'${int(x/1000)}K')
```

```python
import seaborn as sns
```

```python
sns.displot(df_DA_ind['salary_year_avg'],kind='hist')
plt.xlim(0,300000)
ax=plt.gca()
ax.xaxis.set_major_formatter(lambda x,pos : f'${int(x/1000)}K')
```
