```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df.info()
```

```python
df['job_posted_month']=df['job_posted_date'].dt.month_name()
df
```

```python
month=df['job_posted_month'].unique()
dict_month={month:df[df['job_posted_month']==month] for month in month}
dict_month['June']
```

```python
df_q1=pd.concat([dict_month['January'],dict_month['February'],dict_month['March']],ignore_index=True)
df_q1
```

```python
df_q1['job_posted_month'].value_counts().plot(kind='bar')
```
