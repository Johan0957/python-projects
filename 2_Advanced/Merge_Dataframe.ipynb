```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df.info()
```

```python
df_soft=pd.read_csv("https://lukeb.co/software_csv",index_col='job_posted_month')
df_soft
```

```python
df_us=df[df['job_country']=='India'].copy()
df_us
```

```python
df_us['job_posted_month']=df_us['job_posted_date'].dt.strftime('%B')
df_us
```

```python
df_us_pivot=df_us.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_us_pivot.reset_index(inplace=True)
```

```python
df_us_pivot['month_no']=pd.to_datetime(df_us_pivot['job_posted_month'],format='%B').dt.month
df_us_pivot.sort_values(by='month_no',inplace=True)
df_us_pivot.set_index('job_posted_month',inplace=True)
df_us_pivot.drop('month_no',axis=1,inplace=True)
```

```python
df_us_pivot
```

```python
df_us_pivot.merge(df_soft,on='job_posted_month')
top_jobs=df_us_pivot.sum().sort_values(ascending=False).head(5).index.to_list()
top_jobs
```

```python
df_us_pivot[top_jobs].plot(kind='line')
plt.legend()
plt.show()
```
