```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df.info()
```

```python
df_us=df[df['job_country']=='India'].copy()
df_us
```

```python
df_us['job_posted_month']=df_us['job_posted_date'].dt.strftime('%B')
df_us
```

```python
df_us_pivot=df_us.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_us_pivot.reset_index(inplace=True)
```

```python
df_us_pivot['month_no']=pd.to_datetime(df_us_pivot['job_posted_month'],format='%B').dt.month
df_us_pivot.sort_values(by='month_no',inplace=True)
df_us_pivot.set_index('job_posted_month',inplace=True)
df_us_pivot.drop('month_no',axis=1,inplace=True)
```

```python
df_us_pivot.plot(kind='line')
plt.tight_layout()
```

```python
top_3=df_us['job_title_short'].value_counts().head(3).index
df_us_pivot=df_us_pivot[top_3]
```

```python
df_us_pivot.plot(kind='line')
plt.title('Monthly Job Posting For Top Data Jobs In India')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.show()
```
